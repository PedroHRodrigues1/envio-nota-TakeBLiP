```python
import pandas as pd

nota = pd.read_csv("Nota de Atendimento 01-02 até 16-02.csv")

nota = nota.drop(["Total", "%"], axis=1)

display(nota)
```

```python
nota[["Ticket", "Atendente", "Fila de Atendimento", "Nota de Atendimento"]] = nota["Ações"].str.split('|', expand=True)
nota
```

```python
nota = nota.drop("Ações", axis=1)
display(nota)
```

```python
nota["Nota de Atendimento"] = pd.to_numeric(nota["Nota de Atendimento"], errors="coerce")

nota = nota.dropna(how="all", axis=1)
nota = nota.dropna(how="any", axis=0)
# nota = nota.groupby("Fila de Atendimento").mean()["Nota de Atendimento"]
# display(nota)

print(nota.info())
```

```python
# nota[nota["Nota de Atendimento"] == "3"]
# nota[nota['Nota de Andimento'].isin(['10.0'])]
# nota[nota['Nota de Atendimento'].str.len() > 9]
 
nota_boleto = nota[nota["Fila de Atendimento"].str.contains("Boleto")]
# nota


display(nota_boleto)
```

```python
nota_boleto.to_excel('Nota Boleto 16-02.xlsx', sheet_name='Nota 16-02', header=True, index=False)
```

```python
nota_atendimento = nota[nota["Fila de Atendimento"].str.contains("Atendimento")]
display(nota_atendimento)
```
